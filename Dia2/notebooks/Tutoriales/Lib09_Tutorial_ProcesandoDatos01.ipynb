### Asignacion de variables en Python

Si notaron en el codigo anterior, hemos asignado los valores a las variables `nombre, voto` en una sola linea. En lugar de asignar valores por separado a cada elemento, podemos asignarlos al mismo tiempo usando una tecnica que se llama `asignacion multiple`. La linea :

    nombre, voto = partes
    
asigna a cada variable el elemento correspondiente en la lista. 

Tomen un momento para jugar con esta idea de asignacion multiple en IPython Notebook escribiendo instrucciones como las siguientes:

    a, b, c = [1, 2, 3]
    print(b)
    print(a)

Y las siguientes:

    name, cheese, cracker = "Fred,Jarlsberg,Rye".split(",")
    print(cheese)

Escriban sus observaciones:

Es probable que hayan notado que las asignacion multiple de variables unicamente funciona cuando el numero de elementos que se van a asignar del lado izquiero del " = " es igual al numero del lado derecho.

Por ejemplo, la siguiente asignacion no funciona:

    x, y, z = [1, 2]

Tampoco esta:

    nombre, queso, galleta = "Sara,Manchego,Rye,Gamesa".split(",")

Como pueden ver en la documentacion de `split()`, puden pasar un segundo argumento, opcional, que indica el maximo numero de veces que se aplicara `split` a la cadena. Esto puede usarse para prevenir errores al crear demasiados elementos:

    nombre, queso, galleta = "Sara,Manchego,Rye,Gamesa".split(",",2)
    print(nombre)
    print(galleta)

Intentenlo!

Nota: el argumento opcional es *el numero de veces que se aplicara `split` a la cadena,* **no** el numero de partes en las que se dividira. Asi, separando usando `split` una vez crea dos cadenas, separando dos veces crea tres cadenas, etc.

# Factorizando nuestro codigo

Nuestro programa comienza a hacerse un poco largo y complejo. Puede tener muchos comentarios explicando que realiza una o varias secciones especificas, etc.

Puede ser util separar en funciones cada parte, esto hara que sea mas facil de leer:

* Una funcion que haga limpieza de cadenas para que sea mas facil comparar y encontrar iguales (hacemos exactamente lo mismo cuando limpiamos los nombres de rabanos o los nombres de votantes);
* Una funcion para checar si alguien ya ha votado;
* Una funcion para contar el voto.

Intenten separando partes de su codigo en funciones, una por una sin deshacer la funcionalidad del programa.

En IPython Notebook esto es muy sencillo porque pueden ejecutar la celda despues de cada cambio para ver si el programa sigue funcionando.

```python
# Escriban y ejecuten el codigo en esta celda

# Create an empty dictionary for associating radish names
# with vote counts
counts = {}

# Create an empty list with the names of everyone who voted
voted = []

# Clean up (munge) a string so it's easy to match against other     strings
def clean_string(s):
    return s.strip().capitalize().replace("  "," ")

# Check if someone has voted already and return True or False
def has_already_voted(name):
    if name in voted:
        print(name + " has already voted! Fraud!")
        return True
    return False

# Count a vote for the radish variety named 'radish'
def count_vote(radish):
    if not radish in counts:
        # First vote for this variety
        counts[radish] = 1
    else:
        # Increment the radish count
        counts[radish] = counts[radish] + 1


for line in open("../encuestarabanos.txt"):
    line = line.strip()
    name, vote = line.split(" - ")
    name = clean_string(name)
    vote = clean_string(vote)
    
    if not has_already_voted(name):
        count_vote(vote)
    voted.append(name)

print("Results:")
print()
for name in counts:
    print(name + ": " + str(counts[name]))
```

Les parece que el codigo es mas facil de entender en esta forma "modular"?

Para programas pequeños no es muy importante, pero para programas grandes puede ser muy importante hacerlo modular de modo que sea facil de entender y que pueda reusarse.
