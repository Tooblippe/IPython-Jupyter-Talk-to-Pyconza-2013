# Preparar, analizar y presentar datos
Los datos recien adquiridos por el sensor no deberian necesitar preparacion (buscar valores no definidos, errores, etc.). Dado que son datos muy triviales no realizaremos ningun analisis sino unicamente los vamos a graficar para presentarlos.

Como las coordenadas geograficas son pares de puntos, podemos graficarlos directamente:

```python
acel = sensorData.iloc[:,0:3]
```

```python
acel.plot(title='Aceleracion al caminar')
```

Vamos a convertir las coordenadas del vector aceleracion en cada instante de tiempo registrado a un valor escalar (un solo numero) calculando lo que se llama la magnitud del vector aceleracion:

```python
mag_acel = acel.values.sum(axis=1)
```

```python
plt.plot(mag_acel, 'b-')
plt.xlabel('Tiempo (s)')
plt.ylabel('Aceleracion m/s^2');
```

Vamos a eliminar los efectos de la gravedad restando el promedio:

```python
mag_acel_nomean = mag_acel - mag_acel.mean()
```

```python
plt.plot(mag_acel_nomean, 'b-')
plt.xlabel('Tiempo (s)')
plt.ylabel('Aceleracion m/s^2');
```

De este modo los datos estan centrados alrededor del cero y podemos ver claramente que existen picos en la magnitud de la aceleracion. Cada pico corresponde a un paso. 

Incluso podemos ver que hay algunas ligeras correlaciones entre picos y valles bajando (rojo) y subiendo (azul).
