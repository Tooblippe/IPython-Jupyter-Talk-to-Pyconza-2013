# Buscando contornos

Primero vamos a cortar la imagen

```python
img1 = img[750:1250,1000:1550].copy()
plt.imshow(img1)
```

```python
plt.imshow(img1,cmap='gray')
```

```python
bajo = 0.1
alto = 0.8
oscuro = measure.find_contours(img1, bajo)
claro = measure.find_contours(img1, alto)
```

```python
plt.imshow(img1, cmap='gray')

for n, contour in enumerate(oscuro):
        plt.plot(contour[:, 1], contour[:, 0], c='r', linewidth=1)

plt.axis('image')
plt.xticks([])
plt.yticks([]);
```

```python
plt.imshow(img1, cmap='gray')

for n, contour in enumerate(claro):
        plt.plot(contour[:, 1], contour[:, 0], c='r', linewidth=1)

plt.axis('image')
plt.xticks([])
plt.yticks([]);
```
