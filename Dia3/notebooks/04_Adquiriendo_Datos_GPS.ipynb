# Preparar, analizar y presentar datos
Los datos recien adquiridos por el sensor no deberian necesitar preparacion (buscar valores no definidos, errores, etc.). Dado que son datos muy triviales no realizaremos ningun analisis sino unicamente los vamos a graficar para presentarlos.

Como las coordenadas geograficas son pares de puntos, podemos graficarlos directamente:

```python
%matplotlib inline

latlon = sensorData[['Lat','Long']]
latlon.plot(x='Lat', y='Long', kind='scatter')
```

No se ve muy interesante...estos son datos de coordenadas geograficas asi que quizas conviene usar un sistema apropiado...les presento `Basemap` una librearia para analizar datos geograficos. Para instalarla ejecuten el siguiente comando:
    
    conda install basemap

Si todo sale bien durante el proceso de instalacion deben ser capaces de importar la libreria escribiendo el siguiente comando en una celda:

    from mpl_toolkits.basemap import Basemap
    

```python
from mpl_toolkits.basemap import Basemap
from datetime import datetime
```

```python
lats = latlon['Lat']
lons = latlon['Long']

timeStamp = sensorData['Timestamp']
```

```python
m = Basemap(projection='hammer',lon_0=-90)

x, y = m(lons.values,lats.values)

m.drawmapboundary(fill_color='#99ffff')
m.fillcontinents(color='#cc9966',lake_color='#99ffff')
m.scatter(x,y,24,marker='o',color='k')
plt.text(x[-1],y[-1],'LANIA',color='b');
```

Ahora se ve un poco mejor! Podemos usar una proyeccion diferente:

```python
map = Basemap(projection='mill',lon_0=-90)
# plot coastlines, draw label meridians and parallels.
map.drawcoastlines()
map.drawparallels(np.arange(-90,90,30),labels=[1,0,0,0])
map.drawmeridians(np.arange(map.lonmin,map.lonmax+30,60),labels=[0,0,0,1])
# fill continents 'coral' (with zorder=0), color wet areas 'aqua'
map.drawmapboundary(fill_color='#99ffff')
map.fillcontinents(color='coral',lake_color='#99ffff')
x, y = map(lons.values,lats.values)
map.scatter(x,y,24,marker='o', color='b')
plt.text(x[-1],y[-1],'LANIA',color='b');
```

Podemos incluir la informacion temporal (obtenida con el sensor):

```python
map = Basemap(projection='mill',lon_0=-90)
# plot coastlines, draw label meridians and parallels.
map.drawcoastlines()
map.drawparallels(np.arange(-90,90,30),labels=[1,0,0,0])
map.drawmeridians(np.arange(map.lonmin,map.lonmax+30,60),labels=[0,0,0,1])
# fill continents 'coral' (with zorder=0), color wet areas 'aqua'
map.drawmapboundary(fill_color='#99ffff')
map.fillcontinents(color='coral',lake_color='#99ffff')
x, y = map(lons.values,lats.values)
map.scatter(x,y,24,marker='o', color='b')
plt.text(x[-1],y[-1],'LANIA',color='b');
# date = datetime.utcnow()
date=datetime.utcfromtimestamp(timeStamp.values[-1])
CS=map.nightshade(date)
plt.title('Day/Night Map for %s (UTC)' % date.strftime("%d %b %Y %H:%M:%S"));
```

Ahora copien el codigo de la celda que genero el ultimo mapa en una nueva celda y observen los cambios.

Veamos un poco mas de cerca donde esta LANIA:

```python
map = Basemap(projection='mill',lon_0=-90, lat_0=0,
    llcrnrlon=-120, llcrnrlat=0.0,
    urcrnrlon=-70, urcrnrlat=30.75)
# plot coastlines, draw label meridians and parallels.
map.drawcoastlines()
map.drawparallels(np.arange(map.latmin,map.latmax,10),labels=[1,1,1,1])
map.drawmeridians(np.arange(map.lonmin,map.lonmax,20),labels=[0,0,0,1])
# fill continents 'coral' (with zorder=0), color wet areas 'aqua'a
map.drawmapboundary(fill_color='#99ffff')
map.fillcontinents(color='coral',lake_color='#99ffff')
x, y = map(lons.values,lats.values)
map.plot(x, y, 'bo', markersize=18)
plt.text(x[-1],y[-1],'LANIA');
```
