## Analisis de datos usando la libreria Pandas
`pandas` es un paquete de python que proporciona acceso a estructuras de datos de manera rapida, flexible y de sencilla expresion, disenada para trabajar con datos que tienen atributos o tablas relacionales, de manera simple e intuitiva.

Los dos tipos principales de estructuras de datos que emplea `pandas`, `Series` (uni-dimensionales) y `DataFrame` (bi-dimensionaes), son suficientes para manipular la gran mayoria de casos tipicos de analisis de datos en diversas areas como  finanzas, estadistica , ciencias sociales, ingenierias, etc.

<!---
For R users, DataFrame provides everything that R’s data.frame provides and much more. pandas   is built on top of NumPy and is intended to integrate well within a scientific computing environment with many other 3rd party libraries.

```python
import pandas as pd
%matplotlib inline
```

Usaremos datos climaticos de Xalapa correspondientes a las ultimas 24 horas, obtenidos [de este sensor de la Comision Nacional del Agua](http://smn.cna.gob.mx/emas/catalogo/XALAPAob.html)

```python
archivo = './VR23_10M.xls'
pd.read_excel(archivo).head()
```

```python
pd.read_excel(archivo, header=4).head()
```

```python
pd.read_excel(archivo, header=4, index_col='FECHA').head()
```

Conviene asignar una variable a esta accion:

```python
xalapaOb =pd.read_excel(archivo, header=4, index_col='FECHA', parse_dates=True, encoding='utf-8')
xalapaOb.head()
```

Veamos los encabezados de cada columna:

```python
xalapaOb.columns
```

```python
xalapaOb.columns[0]
```

```python
xalapaOb.columns[0].strip()
```

Necesitamos remover los espacios en blanco de los encabezados, de modo que podamos acceder a ellos de manera simple. Es decir en lugar de

```python
xalapaOb[[u'      DIRS']].head()
```

```python
xalapaOb[[u'      TEMP']].head()
```

queremos acceder a las columnas usando 

        xalapaOb[['TEMP']]

Que pasa si ejecutan esta instruccion?


Para renombrar columnas usamos el metodo `rename` de la estructura:

```python
xalapaOb.rename(columns=lambda x: x[:].strip(), inplace=True)
xalapaOb.rename(columns = {'RAD-SOL':'RADSOL'}, inplace=True)
xalapaOb.columns
```

Podemos extraer datos facilmente, por ejemplo la temperatura entre las 3:00 y las 12:00 horas. Para esto es util tener una referencia del indice de la estructura:

```python
madrugada = xalapaOb['2014-10-03 3:00:00' : '2014-10-04 6:00:00']
madrugada.TEMP.head()
```

Veamos las horas de madrugada en las que la temperatura bajo de la media del dia:

```python
madrugada[madrugada.TEMP > media].head()
```
