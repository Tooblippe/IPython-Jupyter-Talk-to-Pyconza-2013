Exploremos algunas construcciones más complejas, por ejemplo usando bucles (loops):

```python
linea = '1 2 3 4 5'
campos = linea.split()
campos
```

```python
total = 0
for campo in campos:
    total += int(campo)
total
```

"list comprehension" es una forma compacta de escribir muchos loops:

```python
numeros = [int(campo) for campo in campos]
numeros
```

```python
sum(numeros)
```

Podemos obtener lo mismo con una notación mas compacta:

```python
sum([int(campo) for campo in campos])
```

Incluso podemos reemplazar la variable por la lista:

```python
sum([int(campo) for campo in linea.split()])
```

Noten que al hacer mas y mas compacta la estructura, se vuelve un poco mas dificil de leer.

Now let's take a look at defining and using functions in Python. We will start by defining a function to evaluate a polynomial.

```python
def poly(x, a, b, c):
    y = a*x**2 + b*x + c
    return y
```

```python
x = 1
poly(x, 1, 2, 3)
```

We can also pass an array as the argument x.

```python
x = array([1, 2, 3])
poly(x, 1, 2, 3)
```

And also do this for a 10-element array, and quickly make a plot of the values.

```python
x = arange(10)
plot(x, poly(x, 2, 3, 4))
```

We can also specify keyword arguments, which are specified by name.

```python
def poly(x, a=1, b=2, c=3):
    y = a*x**2 + b*x + c
    return y
```

```python
poly(x)
```

```python
poly(x, b=5)
```

Now lets create a class. In Python, every class should derive from object. Our class will describe a person, with a name and an age. We will supply a constructor, and a method to get the full name.

```python
class Person(object):
    def __init__(self, first, last, age):
        self.first = first
        self.last = last
        self.age = age
    
    def full_name(self):
        return self.first + ' ' + self.last
```

Now we can create an instance of a Person, and work with the attributes of the class.

```python
person = Person('Mertle', 'Sedgewick', 52)
person.first
```

```python
person.last
```

```python
person.full_name()
```

```python
person.last = 'Smith'
```
